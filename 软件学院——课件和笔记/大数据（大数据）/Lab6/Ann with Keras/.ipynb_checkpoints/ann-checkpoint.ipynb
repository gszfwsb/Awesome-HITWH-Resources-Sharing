# Part 1 - Data Preprocessing

```python
# Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset

dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values

# By using specifying index numbers, we exclude some columns, such as row number, customer ID, surname,...
# they are some features (or indpendent varibales) in input datset X

# Question 1: why are they excluded?
# Question2: is there any other column not included in dataset X?

# How to set dataset y? 
# Question 3: Why  is y's index specification [:, 13]?
```

Here, dependent variable (lables or outcome), y, takes binary values 0 or 1, as it is categorical. So it takes numerical values so we don't need to apply a label encoder.

However, we have our independent variables which have categories that are strings and therefore we need to encode the categorical independent variables. They coutries (France, Spain, Germany) and gender (female, male)

Beore we encode them, let's see our matrix of features from the console.


```python
X
```

```python
# convert France, Spain and Germany into numbers 0 1 and 2.

#  Encord female and male into 0 and 1
```

```python
# Encoding categorical data

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# you may look at the specification of the two imported classes, by removing "#" from the following two comments:

# LabelEncoder?

# OneHotEncoder?


labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```python
X
```

```python

# We are ready to split the data set into training set and the test set.

# totally, dataset 10,000 observations (or instances)

# We will change the test size to 0.2 of original dataset
# as a result,  we have 8000 observations as train data and 2000 observations as test data
```

```python
# Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```


Now time for feature Scalling. Do we need this in general deep learning? The answer is yes absolutely.

That is thoroughly compulsory, otherwise it is going to be a lot of confusion in highly intensive calculations.
So we need to apply feature scaling to ease the calculations. 

Inaddition, we don't want to have one independent variable dominating another one. 

So very importantly, we need to apply feature scale. 



```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Part 2 - Building the ANN!


At firs,we need to import Keras libraries

The full document of Keras API is given in https://keras.io


The core data structure of Keras is a model, a way to organize ANN layers. 

The simplest type of model is called Sequential model, a linear stack of layers, which we will use in this sample.

```python
# Importing the Keras libraries and packages

import keras
from keras.models import Sequential
from keras.layers import Dense

```

```python
# You may see my comments at first in following celles
# Then, you may use "help" or "?" to see the specification of a class
# by removing "#', if you want

# help (keras.models)
```


To start with creating our ANN, the step consists of 

- initializing the ANN and

- defining it as a sequence of layers.

There are two ways of initializing a model,either by defining the sequence of layers or defining a graph.
Here, we'll make an ANN with successive layers and so, well will initialize our deepening model as a sequence of layers.



```python
# Before we initializing a mode, let's review an ANN's structure in general
```

This object is to create the neural network model that will play a role of a classifier, 
because our problem is a classification problem where we have to predict a class.

So this neural network model is going to be a classifier. Also, we're going to predict the test results using this classifier.


```python
# Initialising the ANN

classifier = Sequential() # create a Sequential ANN model or network.

# Adding the input layer (and the first hidden layer, later)

classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))

# Dense() has many arguments, only some of them are used in this sample.
# units - the number of neuros or nodes in this adding layer
# kernel_initializer - It draws samples from a uniform distribution within [-limit, limit], 
# with limit = sqrt(3 * scale / n).
# activation - the name of selected activation function 
# input_dim - the number of independent variables, features, (10) + the number of dependent number (1), label, = 11

# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
# Let's review the two activation functions, "relu" and "sigmoid", which were discussed in the lecture
```

```python
# Compiling the ANN

classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# "stochastic great descent" algorithm or optimizer is used as 'adam'.
```


This compiled method contains several parameters.



```python
# The first argument, optimizer = 'adam'
```



An optimizer is the algorithm to find the optimal set of weights in the ANN, because our neural network is built with the different layers and at moment, the weights are only initialized. 

Now, we apply some sort of algorithm to find the best weights. As we discussed in the lecture, "stochastic
great descent" algorithm will be employed in general. It's called "Adam". 

You may see a set of difference "loss" or cost functions with Keras from: https://keras.io/losses/



```python
# The second argument, loss = 'binary_crossentropy'
```


This corresponds to the last function within "the stochastic gradient decent".
If you go deeper into the mathematical details of this algorithm or optimizer, you will see that it is based on a "loss function" or "cost function" to find the optimal weights. 

And so, we have to specify it. For example, one of "cost functions" in linear regressions is the sum of square errors. That is the sum of the square differences between the real value and the predicted value.

In this project, the "loss function" is defined by "binary_crossentropy", which is exressed as below.

Note: 
The formular is more general. In our case, N=1, since we only have one binary outcome, y.

```python
# Fitting the ANN to the Training set
```

So far, we've only built the ANN without making the connection to the training set.

Now we need to make this connection. 

We are going to use the fit method that will fit our ANN with the training sets.




```python
# Fitting the ANN to the Training set

classifier.fit(X_train, y_train, batch_size = 10, epochs = 100) 

# here, we try epoch = 10 at first, as it does not take a long time 
# Originally, we need to set epoches = 100, 
# you may run it with 100 and get more accurate result.

```

```python
# the four arguments as above:
```


The first argument, X_train, is the our training set to train our classifier.
y_train contains the actual outcomes of the dependent variable for all the observations in our trainset.


Besides of them, we will add two arguments, "batch_size" and "epoch".
The batch size is the number of observations or instances after which you want to update the weights.

As lectured, training an ANN consists of applying steps 1 to 6 over many times (epochs). 
Let's review the importants of the steps:



1). Randomly initialize the weights to small numbers close to 0 (but not 0).

2). Input the first instance of the dataset in the input layer; each feature in one input node (or neuro)

3). Forward-propagation: from left to right, the neuros are activated in a way that impact of each neuros activation is limited by the weights. Propagate the activations until getting the predicted result y.
Compare the predicted result to the actual results; Measure the generated error.

4). Back-Propagation: from right to left. The error is backe-propagated. 

5). Update the weights according to how much they responsible for the error. The learning ratedecided by how much we update the weights.

6). Repeat steps 1) to 5) and update the weights after each instance ( reinforcement learning). Or: repeat steps 1) to 5), but update the weights only after a batch of instances (batch learning)

7.) When the whole training set passed through the ANN, that makes an epoch. Redo more epochs.


# Part 3 - Making predictions and evaluating the model

```python

# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)


#  added by zdw
print("See training finished")


# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

y_pred
```
